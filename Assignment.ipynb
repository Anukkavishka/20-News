```python
import sys
import os
from random import shuffle
root='.\\data'
root_content=os.listdir(root)
classes =root_content

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

import re
dic={}
lemm =[]
sentence =""
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
namecounter=1
#################################################################################################
def process_clustered_training(input_content,label) :
    lemm =[]
    sentence =""
    print("Total number of words in training set raw :"+ str(len(input_content)))
    words=re.split('[^a-zA-Z]',input_content)
    stopWords = set(stopwords.words('english'))
    new_word_list=[]
    for w in words:
        if w not in stopWords:
            if(len(w)>1):
                new_word_list.append(w.strip().lower())
    
    wnl = WordNetLemmatizer()
    lemmatize_list = [wnl.lemmatize(w,"v") for w in new_word_list]
    
    for word in lemmatize_list:
        sentence=word+" "+sentence
    
    lemm.append(sentence)
    
    vector = CountVectorizer()
    newvector = vector.fit_transform(lemm)
    features = vector.get_feature_names()
    
    df = pd.DataFrame(newvector.toarray(), columns = features)
    global namecounter
    
    df.to_csv("train_"+str(label)+".csv")
    namecounter+=1
    
    
    ##building the dictionary
    doc_dic={}
    for w in new_word_list:
        if w not in dic :
            dic[w]=1
    print('Number of unique words in train set',len(dic)) 

###############################################################################################



def process_clustered_testing(input_content,label) :
    lemm =[]
    sentence =""
    print("Total number of words in test set raw :"+ str(len(input_content)))
    words=re.split('[^a-zA-Z]',input_content)
    stopWords = set(stopwords.words('english'))
    new_word_list=[]
    for w in words:
        if w not in stopWords:
            if(len(w)>1):
                new_word_list.append(w.strip().lower())
    
    wnl = WordNetLemmatizer()
    lemmatize_list = [wnl.lemmatize(w,"v") for w in new_word_list]
    
    for word in lemmatize_list:
        sentence=word+" "+sentence
    
    lemm.append(sentence)
    
    vector = CountVectorizer()
    newvector = vector.fit_transform(lemm)
    features = vector.get_feature_names()
    
    df = pd.DataFrame(newvector.toarray(), columns = features)
    global namecounter
    df.to_csv("test_"+str(label)+".csv")
    namecounter+=1
    ##building the dictionary
    doc_dic={}
    for w in new_word_list:
        if w not in dic :
            dic[w]=1
    print('Number of unique words in test set',len(dic)) 


###############################################################################################

    
###############################################################################################

for element in classes :
    path=os.path.join(root,element)#this line here appends the class type to the root path
    sub_folders=os.listdir(path)#list of files in each class folder
    #print(sub_folders)
    file_label=os.path.basename(os.path.normpath(path))
            
    for sub_elem in sub_folders :
        sub_path=os.path.join(path,sub_elem)#this line here appends the class type to the root path
        sub_folder_content=os.listdir(sub_path)#list of files in each class folder
        print(sub_path)
        shuffle(sub_folder_content)
        train_list = sub_folder_content[1 : int(len(sub_folder_content) * .7)]
        test_list=sub_folder_content[int(len(sub_folder_content) * .7)+1 :int(len(sub_folder_content))]
        
        print("train list type :"+str(type(train_list)))
        print("test list type :"+str(type(test_list)))
        
        print("train list size :"+str(len(train_list)))
        print("test list size :"+str(len(test_list)))
        file_count=0 
        train_big_one = ""
        test_big_one = ""
        train_label=""
        test_label=""
        for file in train_list: #iterating through each file and printing each class and its files contents
            newpath=os.path.join(sub_path,file)
            content=open(newpath,'r',encoding="utf-8",errors='replace').read()
            train_big_one+=content
            train_label=os.path.basename(os.path.normpath(sub_path))
            
            #print(len(big_one))
            #file_count+=len(content)
    
        for file in test_list: #iterating through each file and printing each class and its files contents
            newpath=os.path.join(sub_path,file)
            content=open(newpath,'r',encoding="utf-8",errors='replace').read()
            test_big_one+=content
            
            #print(len(big_one))
            #file_count+=len(content)
    
    process_clustered_training(train_big_one,file_label)
    train_big_one=""
    
    process_clustered_testing(test_big_one,file_label)
    test_big_one=""
    
        
    print("**************************************new root class************************************************************")
    
```
